```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("flowers.png")
plt.imshow(im)
plt.show()
```

```python
plt.imshow(im[:,:,0])
plt.show()
```

```python
plt.imshow(im[:,:,1])
plt.show()
```

```python
plt.imshow(im[:,:,2])
plt.show()
```

```python
ret0,thresh0 = cv2.threshold(im[:,:,0],0,255,cv2.THRESH_OTSU)
plt.imshow(thresh0, cmap = 'binary')
plt.show()
```

```python
ret1,thresh1 = cv2.threshold(im[:,:,1],0,255,cv2.THRESH_OTSU)
plt.imshow(thresh1, cmap = 'binary')
plt.show()
```

```python
ret2,thresh2 = cv2.threshold(im[:,:,2],0,255,cv2.THRESH_OTSU)
plt.imshow(thresh2, cmap = 'binary')
plt.show()
```

```python
im1= thresh1 - thresh0 - thresh2
ret,thresh = cv2.threshold(im1,0,255,cv2.THRESH_OTSU)
plt.imshow(thresh, cmap = 'binary')
plt.show()
```
